```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'ML1.xlsx'
data = pd.read_excel(file_path)
```

```python
future_years = [2024, 2025, 2026, 2027]
```

```python
X = data[['AÑO']]
y_pensionistas = data['PENSIONISTAS']
y_pension_estim = data['PENSION ESTIMADA']
```

```python
model_pensionistas = LinearRegression()
model_pensionistas.fit(X, y_pensionistas)
```

```python
model_pension_estim = LinearRegression()
model_pension_estim.fit(X, y_pension_estim)
```

```python
X_future = pd.DataFrame({'AÑO': future_years})
predicted_pensionistas = model_pensionistas.predict(X_future)
predicted_pension_estim = model_pension_estim.predict(X_future)
```

```python
predictions = pd.DataFrame({
    'AÑO': future_years,
    'PENSIONISTAS PREDICHOS': predicted_pensionistas,
    'PENSIÓN ESTIMADA PREDICHA': predicted_pension_estim
})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['AÑO'], y_pensionistas, label='Pensionistas Históricos', marker='o')
plt.plot(future_years, predicted_pensionistas, label='Pensionistas Predichos', marker='o', linestyle='--')
plt.title('Evolución de los Pensionistas (2024-2027)')
plt.xlabel('Año')
plt.ylabel('Número de Pensionistas (en miles)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['AÑO'], y_pension_estim, label='Pensión Estimada Histórica', marker='o')
plt.plot(future_years, predicted_pension_estim, label='Pensión Estimada Predicha', marker='o', linestyle='--')
plt.title('Evolución de la Pensión Estimada (2024-2027)')
plt.xlabel('Año')
plt.ylabel('Pensión Estimada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(predicted_pensionistas, predicted_pension_estim, color='red', label='Datos Predichos (2024-2027)')
plt.scatter(y_pensionistas, y_pension_estim, color='blue', label='Datos Históricos')
plt.title('Relación entre Pensionistas y Pensión Estimada (2024-2027)')
plt.xlabel('Número de Pensionistas (en miles)')
plt.ylabel('Pensión Estimada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
predictions = pd.DataFrame({
    'YEAR': future_years,
    'PREDICTED PENSIONERS': predicted_pensionistas,
    'PREDICTED ESTIMATED PENSION': predicted_pension_estim
})
predictions['PREDICTED PENSIONERS'] = predictions['PREDICTED PENSIONERS'].apply(lambda x: f"{x:,.0f}".replace(',', '.'))
predictions['PREDICTED ESTIMATED PENSION'] = predictions['PREDICTED ESTIMATED PENSION'].apply(lambda x: f"{x:.3f}".replace(',', '.'))
print(predictions)
```
