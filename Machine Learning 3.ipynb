```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
ruta_archivo = 'ML3.xlsx'
df = pd.read_excel(ruta_archivo, sheet_name='Brecha de género ')
```

```python
X = df[['AÑO']]
y = df['BRECHA GÉNERO EN %']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
años_para_predecir = np.arange(2022, 2101).reshape(-1, 1) 
predicciones = modelo.predict(años_para_predecir)
```

```python
año_brecha_cero = años_para_predecir[np.where(predicciones <= 0)[0][0]]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(años_para_predecir, predicciones, color='red', linestyle='--', label='Predicción')
plt.axhline(0, color='green', linestyle='-', label='Brecha 0%')
plt.axvline(año_brecha_cero, color='purple', linestyle=':', label=f'Año {año_brecha_cero[0]}')
```
