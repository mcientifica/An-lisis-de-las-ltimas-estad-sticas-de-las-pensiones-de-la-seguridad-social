```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
file_path = 'ML2.xlsx'
data = pd.read_excel(file_path, sheet_name='Edad media 2020 2024')
```

```python
X = data['Año'].values.reshape(-1, 1)
y = data['Edad media'].values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
future_years = np.array([2024, 2025, 2026, 2027]).reshape(-1, 1)
predictions = model.predict(future_years)
```

```python
prediction_df = pd.DataFrame({
    'Año': [2024, 2025, 2026, 2027],
    'Edad media proyectada': predictions
}).round(2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos históricos')
plt.plot(X, model.predict(X), color='green', linestyle='--', label='Línea de tendencia')
plt.scatter(future_years, predictions, color='red', label='Proyecciones futuras')
```

```python
print(prediction_df)
```

```python
full_df = pd.concat([data, prediction_df])
print(full_df.round(2))
```
